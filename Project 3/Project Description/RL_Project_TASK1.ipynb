```python
import gymnasium as gym
from numpngw import write_apng
from IPython.display import Image

# Dont change the map!
map = [
    "SFFFFFFH",
    "FFHHHHFH",
    "FFHHFFFF",
    "FFFHFHHG",
    "FFHHFHHF",
    "FFHFFHFF",
    "FHFFFFFF",
    "FFFFFFFF",
]
```

# Run with graphics

```python
env = gym.make("FrozenLake-v1", desc=map, is_slippery=False, render_mode='rgb_array')
obs = env.reset()

images = [env.render()]
for _ in range(3):
    action = env.action_space.sample()
    obs, reward, terminated, truncated, info = env.step(action)
    images.append(env.render())

write_apng("anim.png", images, delay=250)  # 250 ms between frames
Image(filename="anim.png")
```
